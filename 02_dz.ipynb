```python
# - Вводим с консоли число от 0 до 1000. Выводим меньшую и большую цифру числа.
while(True):
    inp = input('Введите число от 0 до 1000:  ')
    if int(inp) and int(inp)<=1000 :
        break
    else:
        print('че-то не то ввел. еще разок')
a = [int(i) for i in inp]
print('меньшая цифра: ', min(a))
print('большая цифра: ',max(a))
```

```python
# - Нужно хранить соответствие месяцу порядкового номера от 0 до 11. 
# С консоли вводим либо номер месяца от 0 до 11 => получаем название месяца, 
# вводим название месяца => получаем порядковый номер месяца
month_num = dict()
month_num = { 
    'Январь': 1,
    'Февраль':2,
    'Март':3,
    'Апрель':4,
    'Май':5,
    'Июнь':6,
    'Июль':7,
    'Август':8,
    'Сентябрь':9,
    'Октябрь':10,
    'Ноябрь':11,
    'Декабрь':12}
num_month = {v: k for k, v in month_num.items()}

while(True):
    inp = input('Введите номер месяца или его название:  ')
    if inp.isdecimal() and int(inp)<=12:
        n=int(inp)
        print(num_month[n])
        break
    if inp in  month_num.keys():
        print(month_num[inp])
        break
    else:
        print('че-то не то ввел. еще разок')
```
